```python
# I had the idea of finding out how much it would cost to stay at these listings based off the price per night and how many
# minimum number of nights my stay had to be just to see how expensive the range is. 

clean_df.insert(clean_df.columns.get_loc('minimum_nights') + 1, 'total_cost', clean_df['price'] * clean_df['minimum_nights'] ) 
clean_df
```

```python
price_values = clean_df['price'].values # I am trying to find the prices per night to see if there are any potential outliers
# I also want to see how high and how low some people set their prices
price_values
```

```python
custom_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]  # Define custom bin edges
plt.hist(price_values, bins=custom_bins)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show
```

```python
# Gather the highest and lowest values to see the range of our values so we can set a more accurate custom_bins
lowest_value = np.min(price_values) 
highest_value = np.max(price_values)
print("Lowest Value:", lowest_value)
print("Highest Value", highest_value)
```

```python
# apparently according to our check above of the highest and lowest values there is a listing in this data set with the price 
# per night of 0??? I would love to stay there! So lets check which one it is and also how many there are. Also apparently there
# is a listing that costs 51561 DOLLARS???? Wow that is just insane. I would never in a million years spend that much to stay
# at a hotel.
free_listings = clean_df[(clean_df['price'] == lowest_value)]
expensive_listings = clean_df[(clean_df['price'] == highest_value)]
free_listings
```

```python
expensive_listings
```

```python
# As we can see below keeping a lot of these higher values creates a very skewed histograph, which means we must remove some of them
# We can do this by seeing what a general average of all the values are. 

custom_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000]  # Define custom bin edges
plt.hist(price_values, bins=custom_bins)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show
```

```python
# In order to remove the outliers we want to see what a general average before and after the values will be. As well as we want
# to do a frequency distribution of all the values to see which ones we can remove meaning which ones that do not occur very 
# often
average_value = clean_df['price'].mean()

value_counts = np.bincount(price_values)
total_count = len(price_values)
value_percentages = (value_counts / total_count) * 100

print(total_count)
for price, count, percentage in zip(range(len(value_counts)), value_counts, value_percentages):
    print(f"Price: {price}, Count : {count}, Percentage: {percentage:.2f}%")
```

```python
# Unfortunately the code above also includes every single value inbetween the prices we are trying to find
# So in order to only include the prices and nothing in between we will have to rework our code. 
prices, counts = np.unique(price_values, return_counts=True)
print(total_count)
counter = 0

# With the code below we can see each indiviual price, how many times that price occurs and what the frequency of that price is
# Throughout the dataset. 
for price, count in zip(prices, counts):
    percentage = (count / total_count) * 100
    print(f"Price: {price}, Count: {count}, Percentage: {percentage: .3f}%")
    counter = counter + 1 
print(counter)
```
